# Print type of DataStructures in Python Using Print Command

## DataStructure in Python are as follows: 

## Tuples , List , Dictionary , Sets

### Example:

```python
###Creating Tuple and printing the data type using print()
tup1=('Data', "Structure", 1 , True)
print("The Data Structure of tup1 is:",type(tup1), '\n \t \t \t \t', " Value  :", tup1)

###Creating List and printing the data type using print()
list1=[11,22,33,44,55,66]
print("The Data Structure of list1 is:",type(list1), '\n \t \t \t \t', " Value  :", list1)

###Creating Dictionary and printing the data type using print()
dict1={'Python':6,'Analytic':8,'Sql':3}
print("The Data Structure of dict1 is:",type(dict1), '\n \t \t \t \t', " Value  :", dict1)

###Creating set and printing the data type using print()
s1={'a',1,True,False}
print("The Data Structure of s1 is:",type(s1), '\n \t \t \t \t', " Value  :", s1)
```

# Print command Parameters in Python

## The following are the parameters of Print Command in Python

## Object(s) , Seperator , End , File and Flush.

## Object: Any object, and as many as you like. Will be converted to string before printed.

### Example:

```python
obj=5
b=obj
print(obj,b)
```

## Seperator('sep'):Specify how to separate the objects, if there is more than one.

### Example:

```python
list1 = [11,22,33,44,55,66]
print(list1[0],list1[1],list1[2],list1[3],list1[4],list1[5],sep="-")

print('Great','Learning', sep=' ')
```

## End :Specify what to print at the end. 

### Example:

```python
list_ranks = [1,2,3,4,5,6]
for num in list1:
    print(num,end=' Rank\n')
```

## File:An object with a write method.

### Example:

```python
sourceFile = open('python.txt', 'w')
print('Data Structures Using Python', file = sourceFile)
sourceFile.close()
```

## Flush: A Boolean, specifying if the output is flushed (True) or buffered (False). Default is False

### Example:

```python
import sys
import time
  
for i in range(10):
    print(i,end=' ',flush=True)
    time.sleep(1)
```
